### Rule-Based (Sentiment) Analysis
It uses a set of predefined rules that connect words to emotions or attitudes. E.g., Sad -> Negative, Great -> Positive.
A polarity score is used to determine the range of the effect (be it positive or negative). Sad can be -0.7, Great can be 0.8, with neutral at 0. This means polarity ranges from -1 to 0 to 1 (-1.0.1). However, this analysis finds it difficult to handle sarcasm or irony.

We can use both Textblob and Vader, but vader is more verbose in its response as it shows more polarity properties and the compound scores.


```python
sentence_1 = "i had a great time at the movie it was really funny"
sentence_2 = "i had a great time at the movie but the parking was terrible"
sentence_3 = "i had a great time at the movie but the parking wasn't great"
sentence_4 = "i went to see a movie"
```

```python
# Text Blob
from textblob import TextBlob
score_1 = TextBlob(sentence_1)
print(sentence_1, "==>", score_1.sentiment)
score_2 = TextBlob(sentence_2)
print(sentence_2, "==>", score_2.sentiment.polarity)
score_3 = TextBlob(sentence_3)
print(sentence_3, "==>", score_3.sentiment.polarity)
score_4 = TextBlob(sentence_4)
print(sentence_4, "==>", score_4.sentiment.polarity)
```

```python
# Vader
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
vader_sentiment = SentimentIntensityAnalyzer()

score_1 = vader_sentiment.polarity_scores(sentence_1)
print(sentence_1, "\n", score_1, "\n")
score_2 = vader_sentiment.polarity_scores(sentence_2)
print(sentence_2, "\n", score_2, "\n")
score_3 = vader_sentiment.polarity_scores(sentence_3)
print(sentence_3, "\n", score_3, "\n")
score_4 = vader_sentiment.polarity_scores(sentence_4)
print(sentence_4, "\n", score_4, "\n")
```

```python
import transformers
from transformers import pipeline

sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
score_1 = sentiment_pipeline(sentence_1)
print(sentence_1, "\n", score_1, "\n") # Transformer models are confident. They return a label and a sentence score.
score_2 = sentiment_pipeline(sentence_2)
print(sentence_2, "\n", score_2, "\n")
score_3 = sentiment_pipeline(sentence_3)
print(sentence_3, "\n", score_3, "\n")
score_4 = sentiment_pipeline(sentence_4)
print(sentence_4, "\n", score_4, "\n") # Transformer will mis-classify as positive instead of neutral
```

```python
specific_model_pipeline = pipeline("sentiment-analysis", model="finiteautomata/bertweet-base-sentiment-analysis") # Over 600mb
```

```python
score_1 = specific_model_pipeline(sentence_1)
print(sentence_1, "\n", score_1, "\n")
score_2 = specific_model_pipeline(sentence_2)
print(sentence_2, "\n", score_2, "\n")
score_3 = specific_model_pipeline(sentence_3)
print(sentence_3, "\n", score_3, "\n")
score_4 = specific_model_pipeline(sentence_4)
print(sentence_4, "\n", score_4, "\n")
```
