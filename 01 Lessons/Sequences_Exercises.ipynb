## List Comprehension

````python
comprehension = [[i + j for i in range(2)] for j in range(5)]
print(comprehension)
comprehension = [i + j for i in range(2) for j in range(5)]
print(comprehension)
````

````python
# Get odd numbers from 1 to 11 and raise them to the power of 3 (single conditional after the iterable)
import math
comprehension = [int(math.pow(num, 3)) for num in range(1, 11) if num % 2 != 0]
print(comprehension)
# Do the same thing but if it's an even number, return null (multi conditional before the iterable)
comp = [int(math.pow(num, 3)) if num % 2 else "null" for num in range(1, 11)]
print(comp)
````

Python List Comprehensions - Exercise #1
Create the following three lists: products_on_sale, sale_prices, quantities. They should contain the following values respectively:

`['Chair_Type_1', 'Chair_Type_2', 'Chair_Type_3', 'Chair_Type_4']`

`[100, 120, 135, 150]`

`[1000, 1500, 1300]`

The following code cell will execute a nested loop that will deliver all possible combinations of the elements from the products_on_sale, sale_prices and quantities lists:

```python
for chair_type in products_on_sale:
    for price in sale_prices:
        for quantity in quantities:
            print ([chair_type, price*quantity])
```
Use a list comprehension to obtain the same output. Store it in a variable called sales_revenue.

````python
products_on_sale = [f"Chair_Type_{num}" for num in range(1, 5)]
sale_prices = [100, 120, 135, 150]
quantities = [1000, 1500, 1300]

sales_revenue = [[chair_type, price*quantity] for chair_type in products_on_sale for price in sale_prices for quantity in quantities]
sales_revenue = [
    [chair_type, price * quantity]
    for chair_type in products_on_sale
    for price in sale_prices
    for quantity in quantities
]
sales_revenue
````

## Lambda Functions
This is used when you don't want to name a function. It's also known as an anonymous function.
`function = lambda x: x ** 2` => first `X` is the parameter, and `x ** 2` is what happens to x and is returned.  It's the equivalent of the function below
```python
def square(x):
    return x ** 2
```
You can also get the result without assigning it to a variable by providing the argument and the lambda function in different parenthesis.
`(lambda x: x**2) (4)`

````python
div = lambda num: print(num / 2)
div(12) # Output is 6
print((lambda num: num ** 2) (8)) # Single argument
print((lambda num, pow: num ** pow) (8,3)) # Multiple arguments
````

````python
(lambda num: print(num * 10)) (7)
# Can also be used for formulas
formula = lambda x: ((135 - x ** 3) ** 4) / ((1 + x) ** 5)
print(formula(3))
````

````python
customer_reviews = [
    "Sam was a great help to me in the store.",
    "The cashier was very rude to me, I think her name was Eleanor.",
    "Amazing work from Saddam!",
    "Sarah was able to help me fine the items I needed quickly",
    "Lucy is such a great addition to the team",
    "Great service from Sara, she found me what I wanted"
]
sarah_reviews = []
# The ? in the pattern means that the character before it (in this case, 'h') is optional. This means the regex will match both "sarah" and "sara".pattern = r"sarah?"
pattern = r"sarah?"
for review in customer_reviews:
    if re.search(pattern, review.lower()):
        sarah_reviews.append(review)

print(sarah_reviews)
````

````python
# Let's find reviews that start with the letter 'S'
pattern = r"^S"
reviews = [review for review in customer_reviews if re.search(pattern, review)]
print(reviews)
````

````python
# Let's find reviews that end with the letter 'm'
pattern = r"m$"
reviews = [review for review in customer_reviews if re.search(pattern, review)]
print(reviews)
````

````python
# You can use the pipe symbol for 'OR'
pattern = r"need|want"
pattern = r"(need|want)ed" # Would also work for needed or wanted
reviews = [review for review in customer_reviews if re.search(pattern, review)]
print(reviews)
````

````python
# Let's find ones that end with punctuations. In our pattern, ^ stands for NOT, \w stands for word chars, \s stands for whitespaces
pattern = r"[^\w\s]$"
reviews = [review for review in customer_reviews if re.search(pattern, review)]
print(reviews)
````

````python
# We can clean up and remove punctuations from our reviews using the same pattern
no_punc_reviews = []
for review in customer_reviews:
    no_punc_string = re.sub(pattern, "", review) # Replace each punctuation with nothing.
    no_punc_reviews.append(no_punc_string)

print(no_punc_reviews)
````

## TOKENIZATION
This involves the breaking of text into smaller units (tokens) for easy analysis. The most common type of tokens are words, but depending on the usecase, they can be sentences, sub-words or even single characters.

````python
# Import nltk and download 'punkt_tab'. It is used to figure out where sentences begin and end

from nltk.tokenize import word_tokenize, sent_tokenize
sentence = "Her cat's name is Luna. Her dog's name is Saul"
try:
    sentences = sent_tokenize(sentence)
except LookupError:
    import nltk
    nltk.download("punkt_tab")
    sentences = sent_tokenize(sentence)

print(sentences)
print(word_tokenize("her cat's name is luna "))
print(word_tokenize(sentence))
print(word_tokenize("her cat's name is luna and her dog's name is max"))

# [nltk_data] Downloading package punkt_tab to /root/nltk_data...
# ["Her cat's name is Luna.", "Her dog's name is Saul"]
# [nltk_data]   Unzipping tokenizers/punkt_tab.zip.
````

## STEMMING & LEMMATIZATION
Stemming is a process of standardizing text. In this process, words are reduced to their base form i.e., `Connected` -> `Connect`, `Connecting` -> `Connect`. It works by chopping off endings or suffixes from words but it's not always accurate as `Studies` can be reduced to `Stud` instead of `Study`. It's essential as it reduces complexity in the dataset and makes it easier to manage.

While stemming chops off the end of a word, lemmatization is more sophisticated, as it uses a predefined dictionary to find the correct base. This way it knows the context of the word and uses that information to reduce it properly. It uses vocabulary and grammer rules to return valid base forms, while stemming often produces crude roots.

To use lemmatization, we need to download a lexical database called WordNet.


````python
from nltk.stem import PorterStemmer, WordNetLemmatizer


stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

connect_tokens = ["connecting", "connected", "connection", "connections", "connectivity", "connect", "connects"]
for token in connect_tokens:
    print(token, ":->", stemmer.stem(token))
    print(token, " ::: ", lemmatizer.lemmatize(token))

learn_token = ["learning", "learn", "learned", "learns", "learn's", "learners", "learner"]
print("\n")
for token in learn_token:
    print(token, ":->", stemmer.stem(token))
    try:
        print(token, ":::", lemmatizer.lemmatize(token))
    except LookupError:
        import nltk
        nltk.download("wordnet")
        print(token, ":::", lemmatizer.lemmatize(token))

# Let's see a stem mistake
print("\n")
tokens = ["likes", "better", "worse"]
for token in tokens:
  print(token, ":->", stemmer.stem(token))
  print(token, ":::", lemmatizer.lemmatize(token))
````
